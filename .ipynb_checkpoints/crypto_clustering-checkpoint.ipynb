### Deliverable 1: Preprocessing the Data for PCA

```python
# Load in the data set.
file = 'crypto_data.csv'
crypto_df = pd.read_csv(file)
crypto_df
```

```python
# Keep all the cryptocurrencies that are being traded.
clean_crypto_df = crypto_df.drop(crypto_df[crypto_df['IsTrading'] == False].index)

shape = clean_crypto_df.shape[0]
print(shape)
clean_crypto_df
```

```python
# Find null values
for column in all_crypto_df.columns:
    print(f'Column {column} has {all_crypto_df[column].isnull().sum()} null values')
```

```python
# Remove rows that have at least 1 null value.
all_crypto_df = all_crypto_df.dropna()
all_shape = all_crypto_df.shape[0]
print(all_shape)
all_crypto_df
```
